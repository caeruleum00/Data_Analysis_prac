### Q) 남녀 구분없이 공통으로 사용되는 이름은?

```python
# 성, 여성별 등록된 이름 횟수를 정리

# 피벗 테이블을 이용해 이름/성별에 따른 등록 횟수를 정리
# index = 'Name', columns = 'Sex', values = 'Number', aggfunc='sum'

name_df = raw.pivot_table(index = 'Name', values = 'Number', aggfunc = 'sum', columns = 'Sex')
name_df

# 비어있는 데이터에 0을 입력
name_df = name_df.fillna(0)
```

- 남자/여자 비율 차이가 적을 수록 --> 성별 구분 없는 이름!

```python
# 남자/여자 이름 등록수 합계를 계산
# 'sum' 컬럼 추가

name_df['Sum'] = name_df['M'] + name_df['F']
name_df
```

```python
# 남자/여자 등록 비율을 계산

name_df['F_ratio'] = name_df['F'] / name_df['Sum']
name_df['M_ratio'] = name_df['M'] / name_df['Sum']

name_df.head()

# 두 비율의 차이를 계산합니다.
# abs() - > 절대값
# name_df['M_F_Gap'] = 

name_df['M_F_Gap'] = abs(name_df['F_ratio'] - name_df['M_ratio'])
```

```python
# 이름이 가장 많이 사용된 수를 기준으로 내림차순으로 정렬
name_df = name_df.sort_values(by =  'Sum', ascending = False)
name_df.head()
```

```python
# 남자/여자 사용비율의 차이가 0.1보다 작은 경우
cond = name_df['M_F_Gap'] < 0.1
name_df[cond]
```

```python
name_df[cond].head(10).index
```

### James, Mary가 가장 대표적인 미국이름일까?

## Q) 가장 대표적인 미국이름은?
- 최근 트렌드에 따른

```python
# unique()를 통해, 기간(raw['YearOfBirth'])에 들어가는 값들을 살펴봅니다.

raw['YearOfBirth'].unique()
```

#### 세대 기준으로 그룹만들기
한 세대 나누는 기준 30년 : 2020년 기준 30년씩 구분
- 1930년대 이전
- 1960년대 이전
- 1990년대 이전
- 2020년 이전

```python
year_class_list = []
for year in raw['YearOfBirth'] :
    if year <= 1930:
        year_class = '1930년 이전'
    elif year <= 1960:
        year_class = '1960년 이전'
    elif year <= 1990:
        year_class = '1990년 이전'
    else:
        year_class = '2020년 이전'
        
    year_class_list.append(year_class)
    
```

```python
year_class_list
```

```python
raw['year_class'] = year_class_list
raw.head()
```

```python
# pivot_table()을 활용

name_period = raw.pivot_table(index = ['Name', 'Sex'], values = 'Number', aggfunc= 'sum', columns = 'year_class')
name_period = name_period.fillna(0)

name_period = name_period.astype(int) # 데이터프레임 타입을 int형으로 변경

name_period.head()
```

#### 전체 컬럼 합계 계산하기
- 모든 컬럼을 하나씩 더하기
- sum()활용하기 :  df.sum(axis = 1)


```python
name_period.head()

name_period['1930년 이전'] + name_period['1960년 이전'] + name_period['1990년 이전'] + name_period['2020년 이전']
```

```python
# sume (axis = 1)을 활용하여 컬럼별 합계를 추가합니다.
             
                #행번호, 컬럼번호
# name_period.iloc[ 0 , 1 ]
name_period['sum'] = name_period.sum(axis = 1)
name_period.head()
```

```python
# 세대별 등록 비율을 계산

for col in name_period.columns : 
    col_new = col + "비율"
    name_period[col_new] = name_period[col] / name_period['sum']

name_period.head()
```

```python
# 이름 사용수 합계, 2020 이전 비율, 1990년 이전 비율 기준으로 내림차순

name_period = name_period.sort_values(by = ['sum', '2020년 이전비율', '1990년 이전비율'],ascending = False)
name_period
```

```python
name_period.index
```

```python
# 인덱스를 column으로 돌려주는 것이 데이터를 컨트롤하기 편하다
name_period = name_period.reset_index()
name_period.head()
```

```python
# 남자이름만 선택해서 살펴보기 # sex->M

cond = name_period['Sex'] == 'M'
name_period[cond].head(10)
```

#### 2020년 이전 비율이 50%이상인 경우를 살펴보자

```python
# 남자의 경우
cond_age = name_period['2020년 이전비율'] > 0.5
cond_sex = name_period['Sex'] == 'M'
cond = cond_age & cond_sex
name_period[cond].head()
```

### 요즘 남자이름 트렌드는
#### Joshua, Jose, Jonathan, Ryan, Jacob 이네요!

#### 그렇다면 예전에는 많이 썼지만 요즘은 잘 사용하지 않는 이름은 무엇일까요?

```python
# 1930년대와 1960년대 사용비율을 합쳐서 70% 이상이지만
# 2020년 이전 비율은 10% 이하인 경우에 해당하는 이름
# 남자의 경우

cond_age = (name_period['1960년 이전비율']+name_period['1930년 이전비율'] >= 0.7) & (name_period['2020년 이전비율'] <= 0.1)
cond_sex = name_period['Sex'] == 'M'
cond = cond_age& cond_sex

name_period[cond].head(5)
```

#### Donald, Joe, Willie, Walter, Harold는 지금은 잘 사용하지 않는 이름인 것 같네요~
